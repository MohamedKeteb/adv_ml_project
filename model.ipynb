Nous pouvons charger tous les paquets et les modèles BERT et ResNet

```python
import torch
from transformers import BertTokenizer, BertModel
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from pycocotools.coco import COCO
from PIL import Image
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.optim as optim

# Device Configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# text Model BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased').to(device)
bert_model.eval()

# Image Model ResNet
resnet_model = models.resnet18(pretrained=True)
resnet_model.fc = nn.Identity()  # Supprimer la couche de classification
resnet_model = resnet_model.to(device)
resnet_model.eval()

# Transformation of the images
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])


```

We can creat a class to handle the COCO dataset

```python
class COCODataset(Dataset):
    def __init__(self, coco_annotations, image_dir, transform, tokenizer):
        self.coco = COCO(coco_annotations)
        self.image_dir = image_dir
        self.transform = transform
        self.tokenizer = tokenizer
        self.image_ids = self.coco.getImgIds()
        self.annotations = self.coco.loadAnns(self.coco.getAnnIds())

    def __len__(self):
        return len(self.annotations)

    def __getitem__(self, idx):
        annotation = self.annotations[idx]
        image_id = annotation['image_id']
        caption = annotation['caption']
        image_info = self.coco.loadImgs(image_id)[0]
        image_path = f"{self.image_dir}/{image_info['file_name']}"

        # download and transform the image
        image = Image.open(image_path).convert('RGB')
        image = self.transform(image)

        # Tokenize and encode the text
        tokenized = self.tokenizer(caption, return_tensors="pt", padding=True, truncation=True, max_length=128)
        tokenized = {key: val.squeeze(0) for key, val in tokenized.items()}

        return image, tokenized, caption
```

```python
# paths
coco_annotations = "dataset/coco/annotations/captions_train2017.json"
coco_images = "dataset/coco/train2017/"

# initialize the data loader
coco_dataset = COCODataset(coco_annotations, coco_images, transform, tokenizer)
```

We can now start defining our first functions for the model

```python
def coco_collate_fn(batch):
    """
    Collate function to handle variable-length text sequences
    """
    images = torch.stack([item[0] for item in batch])  # stack the images
    captions = [item[1] for item in batch]  # Capture textual annotations
    raw_captions = [item[2] for item in batch]  # Capture raw texts (optional, for display)

    # Apply dynamic padding to textual annotations
    input_ids = [caption['input_ids'] for caption in captions]
    attention_mask = [caption['attention_mask'] for caption in captions]

    input_ids_padded = torch.nn.utils.rnn.pad_sequence(input_ids, batch_first=True, padding_value=0)
    attention_mask_padded = torch.nn.utils.rnn.pad_sequence(attention_mask, batch_first=True, padding_value=0)

    padded_captions = {
        'input_ids': input_ids_padded,
        'attention_mask': attention_mask_padded,
    }

    return images, padded_captions, raw_captions
```

We define the dataloader for the batches

```python
subset_size = 20000

# Load a subset of the dataset
coco_sub_dataset = torch.utils.data.Subset(coco_dataset, range(subset_size))
coco_loader = DataLoader(
    coco_sub_dataset,
    batch_size=32,
    shuffle=True,
    collate_fn=coco_collate_fn
)

```

Now we can define the full layer network to project on the latent space


```python
class SharedModel(nn.Module):
    def __init__(self, text_embedding_size, image_embedding_size):
        super(SharedModel, self).__init__()
        self.fc_text = nn.Linear(text_embedding_size, 128)
        self.fc_image = nn.Linear(image_embedding_size, 128)

    def forward(self, text_embeddings, image_embeddings):
        text_mapped = self.fc_text(text_embeddings)
        image_mapped = self.fc_image(image_embeddings)
        return text_mapped, image_mapped
```

Define the Wasserstein distance

```python

def wasserstein_distance(text_embedding, image_embedding):
    # Create uniform (normalized) discrete distributions
    n_text = text_embedding.size(0)
    n_image = image_embedding.size(0)
    uniform_text = torch.ones(n_text, device=text_embedding.device) / n_text
    uniform_image = torch.ones(n_image, device=image_embedding.device) / n_image

    # Calculate the cost matrix (Euclidean distance)
    cost_matrix = torch.cdist(text_embedding, image_embedding, p=2)  # Calculating the Euclidean distance

    # Using the optimal transport algorithm
    wasserstein_dist = torch.sum(torch.matmul(uniform_text.view(-1, 1), uniform_image.view(1, -1)) * cost_matrix)

    return wasserstein_dist
```

Define a function to extract embeddings

```python
def extract_embeddings(batch_images, batch_texts):
    # Extract image embeddings
    batch_images = batch_images.to(device)
    with torch.no_grad():
        image_embeddings = resnet_model(batch_images)  # Embedding of the images
        image_embeddings.requires_grad_()  # Make sure the gradient is calculated for the image

    # Extract text embeddings
    batch_texts = {key: val.to(device) for key, val in batch_texts.items()}
    with torch.no_grad():
        text_embeddings = bert_model(**batch_texts).last_hidden_state.mean(dim=1)  # Embedding texts
        text_embeddings.requires_grad_()  # Make sure the gradient is calculated for the text

    return image_embeddings, text_embeddings
```

Trian the Model

```python
# Initialize share model
shared_model = SharedModel(text_embedding_size=768, image_embedding_size=512).to(device)
optimizer = optim.Adam(shared_model.parameters(), lr=0.001)

# training loops
epochs = 10
for epoch in range(epochs):
    shared_model.train()
    running_loss = 0.0

    for batch_images, batch_texts, _ in coco_loader:
        batch_images = batch_images.to(device)
        batch_texts = {key: val.to(device) for key, val in batch_texts.items()}

        # Extract embeddings
        image_embeddings, text_embeddings = extract_embeddings(batch_images, batch_texts)

        # Switch to model
        text_mapped, image_mapped = shared_model(text_embeddings, image_embeddings)

        # Calculate the Wasserstein distance
        loss = wasserstein_distance(text_mapped, image_mapped)
        print(f"Loss: {loss.item()}")  # print loss

        # Optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{epochs}], Wasserstein Loss: {running_loss/len(coco_loader):.4f}")
```

```python
from torch.utils.data import DataLoader
import torch.nn.functional as F

# Transformer le texte en embedding
def get_text_embedding(text):
    tokenizer_inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)
    tokenizer_inputs = {key: val.to(device) for key, val in tokenizer_inputs.items()}
    with torch.no_grad():
        text_embedding = bert_model(**tokenizer_inputs).last_hidden_state.mean(dim=1)
    return text_embedding

# Transformer une image en embedding
def get_image_embedding(image_path):
    image = Image.open(image_path).convert("RGB")
    image = transform(image).unsqueeze(0).to(device)
    with torch.no_grad():
        image_embedding = resnet_model(image)
    return image_embedding


# Extraire les embeddings partagés pour les images
def map_image_embeddings(coco_loader):
    image_embeddings = []
    image_paths = []  # Stocker les chemins des images pour identifier les correspondances
    with torch.no_grad():
        for batch_images, _, image_ids in coco_loader:
            batch_images = batch_images.to(device)
            batch_embeddings = resnet_model(batch_images)
            shared_embeddings = shared_model.fc_image(batch_embeddings)
            image_embeddings.append(shared_embeddings)
            image_paths.extend(image_ids)
    return torch.cat(image_embeddings, dim=0), image_paths
```

To Input a text and output an image we define a new Wasserstein distance to support tensors

```python
# Fonction pour calculer la distance de Wasserstein
def wasserstein_distance_bis(text_embedding, image_embedding):
    # Créer des distributions uniformes (normalisées)
    n_text = text_embedding.size(0)
    n_image = image_embedding.size(0)
    uniform_text = torch.ones(n_text, device=text_embedding.device) / n_text
    uniform_image = torch.ones(n_image, device=image_embedding.device) / n_image

    # Assurez-vous que les embeddings sont 2D (ajout de la dimension 1 si nécessaire)
    if text_embedding.dim() == 1:
        text_embedding = text_embedding.unsqueeze(0)  # Redimensionner de [n_features] à [1, n_features]
    if image_embedding.dim() == 1:
        image_embedding = image_embedding.unsqueeze(0)  # Redimensionner de [n_features] à [1, n_features]

    # Calculer la matrice de coût (distance euclidienne)
    cost_matrix = torch.cdist(text_embedding, image_embedding, p=2)  # Distance Euclidienne

    # Utiliser l'algorithme de transport optimal
    wasserstein_dist = torch.sum(torch.matmul(uniform_text.view(-1, 1), uniform_image.view(1, -1)) * cost_matrix)

    return wasserstein_dist
```

Now we can input a text

```python
# Texte à rechercher
input_text = "a book"

# Embedding textuel partagé
text_embedding = get_text_embedding(input_text)
text_mapped = shared_model.fc_text(text_embedding)

# Calculer les embeddings partagés des images
image_embeddings, image_paths = map_image_embeddings(coco_loader)

# Initialiser une liste pour les distances de Wasserstein
wasserstein_distances = []

# Calculer la distance de Wasserstein entre le texte et chaque image
for image_embedding in image_embeddings:
    dist = wasserstein_distance_bis(text_mapped, image_embedding)
    wasserstein_distances.append(dist)

# Trouver l'image la plus proche (celle avec la distance minimale de Wasserstein)
closest_image_index = torch.argmin(torch.tensor(wasserstein_distances)).item()

# Afficher le chemin de l'image la plus proche
closest_image_path = image_paths[closest_image_index]
print(f"L'image la plus proche est : {closest_image_path}")
```
